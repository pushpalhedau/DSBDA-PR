```python
Q1=iris['sepal.width'].quantile(0.25)
Q3=iris['sepal.width'].quantile(0.75)
IQR=Q3-Q1
print("IQR(", IQR, ") =", "Q3(", Q3, ") - Q1(", Q1, ")")
```

```python
lower_limit=Q1-(1.5*IQR)
upper_limit=Q3+(1.5*IQR)
lower_limit,upper_limit
```

```python
df_without_outliers=iris[(iris['sepal.width']>lower_limit)&(iris['sepal.width']<upper_limit)]
df_without_outliers
```

```python
ax = sns.boxplot(x=df_without_outliers["sepal.width"])
```

```python
sns.distplot(x=iris['sepal.width'])

sns.distplot(x=df_without_outliers['sepal.width'])
```
